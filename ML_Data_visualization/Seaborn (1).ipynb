## scatter plot



```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# Scatter plot with additional parameters
print("Scatter plot with hue, style, and size")
sns.scatterplot(
    x="total_bill",  # X-axis variable
    y="tip",         # Y-axis variable
    data=tips,       # Data source
    hue="day",       # Color coding by day of the week
    style="time",    # Different markers for Lunch vs Dinner
    size="size",     # Adjust point size based on party size
)

# Display the plot
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

## barplot


```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# Bar plot with additional parameters
sns.barplot(
    x="day",                 # X-axis variable (day of the week)
    y="total_bill",          # Y-axis variable (total bill)
    data=tips,               # Data source
    hue="sex",               # Color coding by gender
    # ci=95,                 # 95% confidence interval for error bars
    errorbar=('ci', 95),
    palette="muted"          # Custom color palette for different sexes
)

# Display the plot
plt.title("Bar Plot: Total Bill by Day")
plt.show()
```

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# Violin plot with additional parameters
sns.violinplot(
    x="day",                 # Categories for the x-axis (days of the week)
    y="total_bill",          # Numerical variable (total bill amount)
    data=tips,               # Data source
    hue="sex",               # Color coding based on gender (Male/Female)
    split=True,              # Split violins for different hues
    palette="pastel"         # Custom color palette for the hue variable
)

# Display the plot
plt.title("Violin Plot: Total Bill by Day and Gender")
plt.show()
```

## kdeplot


```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# KDE plot with two variables
sns.kdeplot(
    x="total_bill",        # Numerical variable for the x-axis
    y="tip",               # Numerical variable for the y-axis
    data=tips,             # Data source
    hue="day",             # Color encoding based on the day of the week
    fill=True,             # Fills the area under the density plot
    cmap="Blues",          # Custom colormap
#     alpha=0.7,             # Transparency level
#     levels=10,             # Number of contour levels
#     thresh=0.1             # Minimum density threshold to display
)

# Display the plot
plt.title("KDE Plot: Distribution of Total Bill and Tip by Day")
plt.show()
```

## regplot


```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Create a regression plot with additional parameters
sns.regplot(
    x="total_bill",               # Independent variable
    y="tip",                      # Dependent variable
    data=tips,                    # Dataset
    fit_reg=True,                 # Enable the regression line
#     scatter_kws={"color": "blue", "alpha": 0.6, "s": 70},  # Scatter customization
#     line_kws={"color": "green", "linewidth": 2},           # Line customization
)

# Add title and axis labels
plt.title("Regression Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

# Display the plot
plt.show()
```
