```python
import pandas as pd
```

```python
file_name = './DataFiles/stock.csv'
```

```python
data = pd.read_csv(file_name) 
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#group by

group_data = df.groupby('stock')['buyPrice'].sum()
```

```python
#sorting

sorted_df = df.sort_values(by='buyPrice',ascending=False)
```

```python
sorted_df.head()
```

```python
#sorting with multiple colums

sort_new_df = df.sort_values(by=['buyPrice','StockHighest','StockLowest'],ascending=[False,False,True])
```
