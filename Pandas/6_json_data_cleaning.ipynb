```python
import pandas as pd
```

```python
file = "./DataFiles/ecom_data.json"
data = pd.read_json(file)
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Delivered Date'] = pd.to_datetime(df['Delivered Date'])
```

```python
df['Order Data'] = pd.to_datetime(df['Order Data'])
df.dtypes
```

```python
df['email'].dropna(inplace=True)
```

```python
df.head()
```

```python
df.sort_values(by='item count',ascending=True,inplace=True)
```

```python
df['Order Data'].fillna(df['Delivered Date'],inplace=True)
```

```python
df.head()
```

```python
df['Delivered'] = df['Delivered'].fillna(df['Order Data'] < df['Delivered Date'] )
```

```python
df.head()
```

```python
new_data = df
```

```python
new_data.head()
```

```python
data = new_data.drop(columns=['id','first_name','last_name','email','Phone','OrderID','Order Data','Delivered','Delivered Date'])
data.head()
```
